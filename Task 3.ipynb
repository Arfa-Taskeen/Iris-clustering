#                 Task 3 Clustering Iris dataset using k-means
#                                     Unsupervised ML

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

## Import the Iris dataset

```python
iris_filepath = 'https://raw.githubusercontent.com/Arfa-Taskeen/Iris-clustering/master/Iris.csv'
iris_data = pd.read_csv(iris_filepath, index_col='Id')
iris_data.head(10)
```

## Finding optimum no. of clusters using Elbow method for classification

```python
X = iris_data.iloc[:, [0,1,2,3]].values
wss =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters = i).fit(X)
    kmeans.fit(X)
    wss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), wss)
plt.title('Elbow method')
plt.xlabel('No. of clusters')
plt.ylabel('Error')
plt.show()
```

##  the optimum clusters is where the elbow occurs.
## the optimal value of k is between 2 and 4, as the elbow-like shape is formed at k = 3 in the above graph.

# Predicting using k-means clustering with k = 3

```python
kmeans3 = KMeans(n_clusters = 3)
model = kmeans3.fit(X)
y = model.predict(X)
print(y)
```

```python
kmeans3.cluster_centers_
```

## Visualizing the 3 clusters that were formed with the optimal k value. 
### k=3

```python
plt.scatter(X[:, 0], X[:, 1], c = y, cmap='rainbow')
plt.scatter(kmeans3.cluster_centers_[:, 0], kmeans3.cluster_centers_[:, 1], c='black', label='centroids')
plt.title('Sepal Length vs Sepal Width')
plt.legend()
plt.show()
```

### Sepal length Vs Sepal Width

```python
plt.scatter(X[:, 2], X[:, 3], c = y, cmap='rainbow')
plt.scatter(kmeans3.cluster_centers_[:, 2], kmeans3.cluster_centers_[:, 3], c='black', label='centroids')
plt.title('Petal Length vs Petal Width')
plt.legend()
plt.show()
```

### Petal Length Vs Petal Width

```python
print(model.labels_)
```

### Cluster Id's are 0,1,2

```python
colors = np.array(['red', 'green', 'blue'])
predicted_Y = np.choose(model.labels_, [1,0,2]).astype(np.int64)
```
